```python
import keras
import tensorflow
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sonar.csv",header=None)
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
df.describe()
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
encoder =LabelEncoder()
y =encoder.fit_transform(y)
y
```

```python
def scaling(var):
    sc = StandardScaler()
    var_scale=sc.fit_transform(var)
    return var_scale
x = scaling(x)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units=100, activation ="relu", kernel_initializer ="uniform",input_dim=60))
model.add(Dense(units=200, activation = "relu", kernel_initializer ="uniform"))
model.add(Dense(units=500, activation = "relu", kernel_initializer ="uniform"))
model.add(Dropout(rate=0.1))
model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=10,epochs=100)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred>0.5)
y_pred
```

```python
error_rate =model.evaluate(x_test,y_pred)
```

```python
error_rate
```

```python
from keras.models import load_model

model.save('my_model.h5',include_optimizer=False)  # creates a HDF5 file 'my_model.h5'
```
